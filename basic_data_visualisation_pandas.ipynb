```python
import pandas as pd
import matplotlib.pyplot as plt
# read csv file
df_population_raw = pd.read_csv(
    r"C:\Users\(...)\(...) - (...)\Desktop\Data Analytics\Py\(...)\project_2_data_visualisation\population_total.csv"
    )
# show the data sample
df_population_raw.head()
```

```python
# drop null values
df_population_raw.dropna(inplace=True)

# reshape the data using .pivot() method
df_pivot = df_population_raw.pivot(
    index="year",
    columns="country",
    values="population"
)
# select some countries and saving the result
df_pivot = df_pivot[['United States', 'Poland', 'China', 'Japan', 'India']]
```

```python
# plot a lineplot 
df_pivot.plot(
    kind='line',
    xlabel='Year',
    ylabel='Population',
    title='Population (1955-2020)',
    figsize=(8,4)
)
```

```python
# select one year (2020)
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
# transpose table
df_pivot_2020 = df_pivot_2020.T
# sort values
df_pivot_2020 = df_pivot_2020.sort_values(by=2020.0, ascending=False)

# plot single bar chart
df_pivot_2020.plot(
    kind='bar',
    color='lightblue',
    xlabel='Country',
    ylabel='Population',
    title='Population (2020)',
    figsize=(5,6)
)
```

```python
# select a few years
df_pivot_sample = df_pivot[df_pivot.index.isin([
    1980, 1990, 2000, 2010, 2020
    ])]

# plot multiple bar chart
df_pivot_sample.plot(
    kind='bar',
    xlabel='Year',
    ylabel='Population',
    title='Population (1980-2020)',
    figsize=(5,6)
)
```

```python
# change column name
df_pivot_2020.rename(columns={2020:'2020'}, inplace=True)

# plot pie chart
df_pivot_2020.plot(
    kind='pie',
    y='2020',
    title='Population (2020) in %'
    )
```

```python
# plot box plot
df_pivot['United States'].plot(
    kind='box',
    color='darkblue',
    ylabel='Population'
)
```

```python
# plot histogram
df_pivot['Poland'].plot(
    kind='box',
    color='darkblue',
    ylabel='Population'
)
```

```python
# plot histogram
df_pivot[['United States', 'Japan']].plot(
    kind='hist'
)
```
